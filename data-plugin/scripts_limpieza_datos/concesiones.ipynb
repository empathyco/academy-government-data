# Limpieza del conjunto de datos de concesiones

```python
import pandas as pd
import numpy as np
```

```python
headers = ['id',
           'id_convocatoria',
           'administracion',
           'departamento',
           'organo',
           'titulo_convocatoria',
           'url_bases_regul',
           'apli_presupuestaria',
           'fecha_concesion',
           'beneficiario',
           'importe',
           'instrumento',
           'ayuda_equivalente',
           'detalles',
           'codigo_bdns',
           'valor_desconocido',
           'id_interna']

juridicas1 = pd.read_csv("juridicas_1.csv", names = headers)
juridicas2 = pd.read_csv("juridicas_2.csv", names = headers)
concesiones = pd.concat([juridicas1,juridicas2]).reset_index(drop = True)
```

```python
concesiones=concesiones.iloc[:,3:16]
```

# Separación beneficiario - CIF/NIF

Tenemos que crear una nueva columna (CIF/NIF) donde introduciremos la primera parte de la columna benificiario, permitiendo buscar por el CIF o el NIF aparte del nombre del beneficiario.

```python
concesiones["CIF/NIF"]=""
```

```python
concesiones.beneficiario=concesiones["beneficiario"].str.split(n=1)
```

```python
concesiones["CIF/NIF"]=concesiones["beneficiario"].str[0]
```

```python
concesiones["beneficiario"]=concesiones["beneficiario"].str[1:]
```

```python
concesiones["beneficiario"]=concesiones["beneficiario"].str[0]
```
